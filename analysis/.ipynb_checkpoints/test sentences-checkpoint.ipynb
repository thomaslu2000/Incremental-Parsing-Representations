```python
def printSens(text, unique_pattern='a regex expression in the string, prints out if it finds one'):
    strings = [s for s in text.split('\n') if len(s) > 0]
    stringIter = iter(strings)
    for s in stringIter:
        tree = Tree.fromstring(s)
        display(tree)
        cats = next(stringIter).replace('cats: ', '')
        if re.search(unique_pattern, cats):
            print('Found Pattern:', cats)
        cats = re.sub(r'(\[\d+)|(\d+\])', '', cats)
        print(cats, '\n', '='*50,'\n')
```
