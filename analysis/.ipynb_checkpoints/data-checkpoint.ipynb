```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
import nltk
from nltk.tree import Tree
import trees
```

```python
test_treebank = trees.load_trees('../data/02-21.10way.clean')
# test_treebank = trees.load_trees('../data/22.auto.clean')
```

```python
tree_strings = [t.linearize() for t in test_treebank]
```

```python
def convertS(s):
    s = re.sub(r'\([^\s^\)^\(]+','(S', s)
    s = re.sub(r'\(S [^\s^\)^\(]+\)', '(null X)', s)
    return s, len(re.findall(r'X', s))
# convertS(tree_strings[0])
```

```python
def searchCount(i):
    return [s for s, c in [convertS(t) for t in tree_strings] if c == i]
```

```python
convertS(tree_strings[0])
```

```python
cats = searchCount(4)
cat_series = pd.Series(cats)
cat_series.value_counts().plot.bar(title='Counts')
plt.show()
```

```python
'(S (null X) (S(null X)) (S (null X) (null X)))' in cats
```
